**Task 1.** Напишіть скрипт, який перевіряє, чи ціле число ділиться на 5 без залишку. Число вводить користувач.

```python
number = int(input('n>>'))

if number % 5:
  print('no')
else:
  print('yes')
```

```python
number = int(input('n>>'))

res = 'no' if number % 5 else 'yes'
print(res)
```

**Task 2.** Напишіть програму, яка визначає, чи належить число діапазону від 1 до 10.


```python
number = int(input('n>>'))

if 1 <= number <= 10:
  print('yes')
else:
  print('no')
```

**Task 6.** Користувач вводить три числа (дохід, витрати, потрібна сума кредиту). Визначте, чи зможе людина сплатити кредит за N місяців.

```python
import math

profit, expenses, amount = float(input('profit>>')), float(input('expenses>>')), float(input('amount>>'))


res = math.ceil(amount / (profit - expenses))
print(res)
```

**Task 7.** Користувач вводить два числа (масу та зріст). Обчисліть індекс маси тіла (ІМТ) та визначте категорію (норма, надмірна вага тощо).

```python
weight, height = int(input('weight (kg)>>')), int(input('height (cm)>>'))

imt = weight / (height / 100) ** 2

if imt <= 18.5:
  print("Недостаточный вес")
elif imt <= 24.9:
  print("Нормальный вес")
elif imt <= 29.9:
  print("Избыточный вес")
else:
  print("Ожирение")
```
