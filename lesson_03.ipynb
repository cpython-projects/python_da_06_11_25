```python
points = int(input('points>>'))


if points < 0 or points > 100:
  print('Error')
elif points <= 49:
  print('F')
elif points <= 59:
  print('E')
elif points <= 65:
  print('D')
elif points <= 75:
  print('C')
elif points <= 86:
  print('B')
else:
  print('A')
```
