```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/cpython-projects/python_da_06_11_25/refs/heads/main/lesson_13_data.csv')
df
```

# STEP 1

```python
df.columns.to_list()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.rename(columns={
  'ID': 'user_id',
  'Name': 'user_name',
  'Age': 'user_age',
  'Salary ($)': 'salary_usd',
  'Hire Date': 'hire_date',
  'Country': 'country',
  'Weight': 'user_weight'
}, inplace=True)
```

```python
df
```

# STEP 2

```python
df.duplicated().sum()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

# STEP 3

```python
df.info()
```

# STEP 4

```python
df.gender.unique()
```

```python
df['gender'] = df.gender.astype('category')
```

```python
df.country.unique()
```

```python
df['country'] = df.country.str.title()
```

```python
df['country'] = df.country.replace({
    'Usa': 'U.S.A.',
    'Uk': 'United Kingdom'
})
```

```python
df
```

```python
df.user_age.unique()
```

```python
df['user_age'] = df.user_age.replace(
    {
        'Thirty': '30',
        'Twenty': '20'
    }
)
```

```python
df['user_age'] = df.user_age.astype('int')
```

```python
df['user_age'] = pd.to_numeric(df.user_age, errors='coerce')
```

```python
df['user_name'] = df.user_name.str.title()
```

```python
def convert_weight(item: str):
  weight = item.lower().strip()

  if 'kg' in weight:
    return float(weight.replace('kg', ''))

  if 'lbs' in weight:
    return float(weight.replace('lbs', '')) * 0.4535

  try:
    return float(weight)
  except:
    return None
```

```python
df['user_weight_kg'] = df.user_weight.apply(convert_weight)
```

```python
df
```

```python
df['hire_date_new'] = pd.to_datetime(df.hire_date, errors='coerce')
```

```python
df
```

```python
from dateutil.parser import parse

df['hire_date_new'] = df.hire_date.apply(parse)
df
```
