**Task 6.** Дано рядок із текстом, що містить хештеги (наприклад, "#data #Python"). Видобудьте їх у список.

```python
text = '#data #Python'
text = text.split('#')
while '' in text:
  text.remove('')


text = list(map(str.strip, text))
text
```

**Task 7.** Дано рядок з текстом. Залиште тільки слова довжиною понад 4 символи.

```python
x = 'aaa aaaa aaaaa bbbbbb nnnnnnnnn'

x = x.split()
res = [word for word in x if len(word) > 4]
print(res)
```

**Task 8.** У рядку замініть усі цифри на символ \*.

```python
text = 'My card number is 1234567890'

for digit in '0123456789':
  text = text.replace(digit, '*')

text
```

```python
text = 'My card number is 1234567890'

for digit in string.digits:
  text = text.replace(digit, '*')

text
```

**Task 9.** У тексті знайдіть усі слова, що складаються лише з великих літер.

```python
text = 'MY name is OLEH'
res = []
for item in text.split():
  if item.isupper():
    res.append(item)
print(*res)
```

**Task 10.** У рядку видаліть усі повторювані слова, зберігши їхній порядок.

```python
text = 'MY is name is OLEH'
res = []
for item in text.split():
  if item not in res:
    res.append(item)
print(*res)
```
