**Task 1.** Дано число (чотиризначне). Перевірити, чи є воно «щасливим квитком».
Примітка: щасливим квитком називається число, в якому при парній кількості цифр сума цифр його лівої половини дорівнює сумі цифр його правої половини.
Наприклад, 1322 є щасливим квитком, тому що 1+3 = 2+2.

````python
number_ticket = list(input('number>>'))
number_ticket
````

````python
number_ticket = list(map(int, number_ticket))
number_ticket
````

````python
if len(number_ticket) % 2:
    print('непарна кількість цифр')
else:
    mid = len(number_ticket) // 2
    left = number_ticket[:mid]
    right = number_ticket[mid:]

    result = 'Lucky' if sum(left) == sum(right) else 'Unlucky'
    print(result)
````

**Task 9.** Інвестор вклав гроші у фонд на 4 роки. Щорічна дохідність фонду була такою:

* **+10%**
* **+20%**
* **–5%**
* **+15%**

Необхідно оцінити **“середній річний темп зростання”**, тобто число, яке показує, як в середньому змінювалася вартість інвестицій за ці роки, незважаючи на коливання.

````python
import statistics
data = [1.1, 1.2, 0.95, 1.15]
result = (statistics.geometric_mean(data) - 1) * 100
print(f'{result:.2f}%')
````

**Task 10.** Середня швидкість для штрафу (однакові ділянки)

На трасі встановлено автоматичний контроль середньої швидкості для ділянки у 60 км. Щоб перевірити, чи перевищував водій ліміт, інспектор поділив маршрут на 6 рівних ділянок (кожна ділянка — однакова відстань) і зафіксував середню швидкість авто на кожній ділянці.

**Дані (швидкість у км/год для кожної ділянки, ділянки рівні):**

```python
speeds = [92, 108, 95, 102, 99, 85]

```

1. Обчислити **одну загальну середню швидкість** по всій контролюваній ділянці (повернути число в км/год, округлити до 1 знака після коми).
2. Вказати, **чи перевищив водій встановлений ліміт (100 км/год)** — відповісти «так» або «ні» і пояснити коротко (1 речення), як це визначили.

````python
import statistics
speeds = [92, 108, 95, 102, 99, 85]
speed_limit = 100  # км/год


avg_speed = statistics.harmonic_mean(speeds)
print(avg_speed)
````
