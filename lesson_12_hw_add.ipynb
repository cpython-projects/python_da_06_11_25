## Опис стовпців

| Стовпець       | Опис                                       |
| -------------- | ------------------------------------------ |
| `show_id`      | Унікальний ідентифікатор тайтлу            |
| `type`         | Тип контенту: `Movie` або `TV Show`        |
| `title`        | Назва фільму або серіалу                   |
| `director`     | Режисер (може бути порожнім)               |
| `cast`         | Акторський склад (рядок з іменами або NaN) |
| `country`      | Країна або країни виробництва              |
| `date_added`   | Дата додавання на Netflix                  |
| `release_year` | Рік виходу                                 |
| `rating`       | Віковий рейтинг                            |
| `duration`     | Тривалість (хвилини або кількість сезонів) |
| `listed_in`    | Жанр або список жанрів                     |
| `description`  | Короткий опис                              |
