```python
numbers = [12, 23, 34, 12, 23, 12, 12, 23, 45, -45, -23, 0]
```

```python
def filter_odd(numbers):
  res = []
  for item in numbers:
    if item % 2 == 0:
        res.append(item)

  return res

filter_odd(numbers)
```

```python
def filter_gt0(numbers):
  res = []
  for item in numbers:
    if item > 0:
        res.append(item)

  return res

filter_gt0(numbers)
```

```python
def is_odd(item):
  return item % 2 == 0

def is_gt0(item):
  return item > 0


def filter_items(numbers, func):
  res = []
  for item in numbers:
    if func(item):
        res.append(item)

  return res

print(filter_items(numbers, is_odd))
print(filter_items(numbers, is_gt0))
```

```python
print(filter_items(numbers, lambda item: item < 0))
print(filter_items(numbers, lambda item: item > 30))
```

```python
prices = ['12$', '23$', '78$', '34', '78$']

def clean_prices(prices):
  for i, item in enumerate(prices):
    prices[i] = float(item.replace('$', ''))

  return prices

clean_prices(prices)
```

```python
prices = ['12$', '23$', '78$', '34', '78$']

def clean_item(item):
  return float(item.replace('$', ''))

list(map(clean_item, prices))
```

```python
prices = ['12$', '23$', '78$', '34', '78$']

list(map(lambda item: float(item.replace('$', '')), prices))
```
