**Task 1.**
Напишіть скрипт, який для п’яти цілих чисел (вводяться користувачем з клавіатури), знаходить мінімум, максимум і середнє арифметичне.

```python
a = int(input('a>>'))
b = int(input('b>>'))
c = int(input('c>>'))
d = int(input('d>>'))
e = int(input('e>>'))
```

```python
print(f'min = {min(a, b, c, d, e)}')
print(f'max = {max(a, b, c, d, e)}')
print(f'avg = {(a + b + c + d + e) / 5}')
```

```python
min_value = min(a, b, c, d, e)
max_value = max(a, b, c, d, e)
avg_value = (a + b + c + d + e) / 5
print(min_value, max_value, avg_value, sep='\n')
```

**Task 2.**
Напишіть скрипт, який повертає результат арифметичних операцій +, -, *, /, // для заданих x і y.

```python
x = int(input('x>>'))
y = int(input('y>>'))

print(f'x + y = {x + y}')
print(f'x - y = {x - y}')
print(f'x * y = {x * y}')
print(f'x / y = {x / y}')
print(f'x // y = {x // y}')
print(f'x % y = {x % y}')
```
