````python
from collections import Counter
text = 'Hello hello HELLO world'

text = text.lower().split()
result = Counter(text)
print(*result.items(), sep='\n')
````

````python
import string
from collections import Counter

text = input("текст>> ")
words = text.split()
without_punctuation = []
for w in words:
    w = w.strip(string.punctuation).lower()
    if w:
        without_punctuation.append(w)
res = Counter(without_punctuation)
print(res)
````

```python
translations = {
     'hello': 'привіт',
     'goodbye': 'до побачення',
     'cat': 'кіт',
     'dog': 'собака'
}
```

````python
translations = {
     'hello': 'привіт',
     'goodbye': 'до побачення',
     'cat': 'кіт',
     'dog': 'собака'
}

word = input('слово>> ').lower().strip()
print(translations.get(word, 'No word'))
````

````python
translations = {
     'hello': 'привіт',
     'goodbye': 'до побачення',
     'cat': 'кіт',
     'dog': 'собака'
}

word = input('слово>> ').lower().strip()

if word in translations:
    print(translations[word])
else:
    for key, value in translations.items():
        if word == value:
            print(key)
            break
    else:
        print('No word')
````

**Task 4.** Дано список клієнтів, кожен клієнт представлений словником з полями: `"id"`, `"ім'я"`, `"вік"`, `"місто"`.

```python
[
    {"id": 1, "ім'я": "Анна", "вік": 25, "місто": "Берлін"},
    {"id": 2, "ім'я": "Іван", "вік": 30, "місто": "Берлін"},
    {"id": 3, "ім'я": "Ольга", "вік": 22, "місто": "Відень"}
]
```

Напишіть скрипт, який формує словник, у якому ключі — міста, а значення — клієнти (список) з цих міст.

````python
data = [
    {"id": 1, "ім'я": "Анна", "вік": 25, "місто": "Берлін"},
    {"id": 2, "ім'я": "Іван", "вік": 30, "місто": "Берлін"},
    {"id": 3, "ім'я": "Ольга", "вік": 22, "місто": "Відень"}
]

result = {}

for item in data:
    city = item['місто']
    if city not in result:
        result[city] = []
    result[city].append(item["ім'я"])

print(result)
````

````python
from collections import defaultdict

data = [
    {"id": 1, "ім'я": "Анна", "вік": 25, "місто": "Берлін"},
    {"id": 2, "ім'я": "Іван", "вік": 30, "місто": "Берлін"},
    {"id": 3, "ім'я": "Ольга", "вік": 22, "місто": "Відень"}
]

result = defaultdict(list)

for item in data:
    city = item['місто']
    result[city].append(item["ім'я"])

print(result)
````

**Task 5.** Дано список працівників, де кожен працівник представлений словником з полями `"ім'я"`, `"відділ"`.

```python
[
    {"ім'я": "Анна", "відділ": "Маркетинг"},
    {"ім'я": "Іван", "відділ": "Продажі"},
    {"ім'я": "Ольга", "відділ": "Маркетинг"}
]
```

Створіть скрипт, який групує працівників за відділами.

````python
data = [
    {"ім'я": "Анна", "відділ": "Маркетинг"},
    {"ім'я": "Іван", "відділ": "Продажі"},
    {"ім'я": "Ольга", "відділ": "Маркетинг"}
]

res = defaultdict(list)

for item in data:
    department = item['відділ']
    res[department].append(item['ім\'я'])

print(res)
````

**Task 6.** Дано словник з даними щодо продажів, де ключі — дні місяця, а значення — сума продажу.

```python
{
    1: 500, 2: 700, 3: 300, 4: 800, 5: 200
}
```

Знайдіть дні з найбільшими відхиленнями від середнього значення.

````python
import statistics
data = {
    1: 500, 2: 700, 3: 300, 4: 800, 5: 200
}

avg = statistics.mean(data.values())
avg
````

````python
d = [abs(value - avg) for key, value in data.items()]
d
````

````python
max_v = max(d)
for i, item in enumerate(d, start=1):
  if item == max_v:
    print(i)

````
