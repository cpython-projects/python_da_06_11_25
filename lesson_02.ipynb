```python
from decimal import Decimal


a = Decimal('0.1')
b = Decimal('0.3')

print(a + a + a)
print(a + a + a == b)
```

```python
a = input('a>')
b = input('b>')

res = Decimal(a) + Decimal(a) + Decimal(a)
print(res)
print(res == Decimal(b))
```

```python
from decimal import ROUND_DOWN, ROUND_FLOOR, ROUND_HALF_DOWN, ROUND_HALF_EVEN, ROUND_HALF_UP, ROUND_UP, ROUND_CEILING
x = Decimal('3.55')

print(x.quantize(Decimal('0.1'), rounding=ROUND_DOWN))
print(x.quantize(Decimal('0.1'), rounding=ROUND_FLOOR))
print(x.quantize(Decimal('0.1'), rounding=ROUND_HALF_DOWN))
print(x.quantize(Decimal('0.1'), rounding=ROUND_HALF_EVEN))
print(x.quantize(Decimal('0.1'), rounding=ROUND_HALF_UP))
print(x.quantize(Decimal('0.1'), rounding=ROUND_UP))
```
