````python
customer_purchases = [
  {"Customer_ID": 1001, "Product": "Laptop", "Category": "Electronics", "Purchase_Date": "2024-02-01"},
  {"Customer_ID": 1002, "Product": "Smartphone", "Category": "Electronics", "Purchase_Date": "2024-02-02"},
  {"Customer_ID": 1003, "Product": "Sofa", "Category": "Furniture", "Purchase_Date": "2024-02-03"},
  {"Customer_ID": 1001, "Product": "Tablet", "Category": "Electronics", "Purchase_Date": "2024-02-10"},
  {"Customer_ID": 1002, "Product": "Chair", "Category": "Furniture", "Purchase_Date": "2024-02-15"},
  {"Customer_ID": 1003, "Product": "Laptop", "Category": "Electronics", "Purchase_Date": "2024-02-18"},
  {"Customer_ID": 1004, "Product": "Desk", "Category": "Furniture", "Purchase_Date": "2024-02-20"},
]
````

Task 1. Визначте безліч унікальних товарів, які були куплені.

````python
result = {item.get('Product') for item in customer_purchases}
print(*result, sep=',')
````

Task 2. Виведіть список унікальних Customer_ID, які купили хоча б один товар із категорії Electronics.

````python
result = {item.get('Customer_ID') for item in customer_purchases if item.get('Category', '').lower() == 'electronics'}
print(*result, sep=',')
````

Task 3. Визначте клієнтів, які купували як в категорії Electronics, так і в категорії Furniture.

````python
result_1 = {item.get('Customer_ID') for item in customer_purchases
          if item.get('Category', '').lower() == 'electronics'}

result_2 = {item.get('Customer_ID') for item in customer_purchases
          if item.get('Category', '').lower() == 'furniture'}

result = result_1.intersection(result_2)
print(*result, sep=',')
````

Task 4. Знайдіть клієнтів, які купували товари лише з однієї категорії.

````python
from collections import defaultdict

customers = defaultdict(set)

for item in customer_purchases:
  cid = item['Customer_ID']
  customers[cid].add(item['Category'])

result = {k for k, v in customers.items() if len(v) == 1}

print(*result, sep='\n')
````

````python
from collections import defaultdict

customers = defaultdict(set)

for item in customer_purchases:
  cid = item['Category']
  customers[cid].add(item['Customer_ID'])

res = set.symmetric_difference(*customers.values())
print(*res, sep='\n')
````

Task 6. Виберіть двох клієнтів і знайдіть перетин їх покупок (які товари вони купили однакові).

````python
user_1 = 1001
user_2 = 1003

user_1_purchases = {item.get('Product') for item in customer_purchases if item.get('Customer_ID') == user_1}
user_2_purchases = {item.get('Product') for item in customer_purchases if item.get('Customer_ID') == user_2}

result = user_1_purchases & user_2_purchases
print(*result, sep=',')
````

Task 7. Для кожного клієнта визначте товари, які він купив, але ніхто інший із покупців не купував.

````python
users = {item['Customer_ID'] for item in customer_purchases}
users
````

````python
for user in users:

  user_purchases = {item['Product'] for item in customer_purchases if item['Customer_ID'] == user}
  users_others = {item['Product'] for item in customer_purchases if item['Customer_ID'] != user}
  print(user)
  print(user_purchases.difference(users_others))
  print('-' * 20)
````

````python
from collections import defaultdict

users_orders = defaultdict(set)

for item in customer_purchases:
  cid = item['Customer_ID']
  users_orders[cid].add(item['Product'])


print(*users_orders.items(), sep='\n')
````

````python
for user in users:
  print(user)
  others = list(users_orders.values())
  others.remove(users_orders[user])
  res = users_orders[user].difference(*others)
  print(res)
  print('-' * 20)
````
