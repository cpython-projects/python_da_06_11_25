Датасет [Retail Sales Data Set of Alcohol and Liquor](https://github.com/cpython-projects/python_da_06_11_25/blob/main/Warehouse_and_Retail_Sales.zip) містить інформацію про **щомісячні продажі алкогольних напоїв** у роздрібних магазинах та через складську систему.

### **Опис змінних**

* **YEAR** — календарний рік, у якому відбулися продажі

* **MONTH** — місяць продажу (від 1 до 12), використовується для аналізу сезонності

* **SUPPLIER** — постачальник алкогольної продукції

* **ITEM CODE** — унікальний ідентифікатор товару

* **ITEM DESCRIPTION** — назва або опис товару (напій, бренд, тип упаковки тощо)

* **ITEM TYPE** — тип алкогольного напою
  (наприклад: Wine, Beer, Spirits, Liquor тощо)

* **RETAIL SALES** — обсяг роздрібних продажів товару за відповідний місяць

* **RETAIL TRANSFERS** — кількість товару, переданого між роздрібними точками або з центрального складу до магазинів

* **WAREHOUSE SALES** — обсяг продажів товару зі складу (оптові або міжскладські операції)

Компанія займається продажем алкогольних напоїв оптом і в роздріб. Існує великий масив даних про **щомісячні продажі окремих товарів**, що охоплює кілька років. Керівництво хоче краще зрозуміти, як сезонність (сезонні коливання) впливає на продажі алкоголю, і на основі цього планувати маркетингові акції та логістику.
